# Lesson 24. One-Way ANOVA &ndash; Conditions, Scope and Limits &ndash; Part 2 

## Example

First, we repeating the steps we took with the `FatRats` data and example in the previous lesson.

Load the data:

```r
library(Stat2Data)
data(FatRats)
```

Create a new data frame, containing only the observations for the rats fed a high-protein diet:

```r
FatRats.HighProtein <- FatRats[FatRats$Protein == 'Hi', ]
```

Perform one-way ANOVA:

<div class="alert alert-info">
    Note the slightly different syntax from Lesson 23:
    <ul>
        <li>A simpler formula <code>Gain ~ as.factor(Source)</code>, in conjunction with the keyword argument <code>data = FatRats.HighProtein</code></li>
        <li><code>as.factor(Source)</code> ensures that R treats the <code>Source</code> variable as categorical</li>
    </ul>
</div>

```r
test <- aov(Gain ~ as.factor(Source), data = FatRats.HighProtein)
summary(test)
```

We compute the standard deviations for each group:

```r
tapply(FatRats.HighProtein$Gain, FatRats.HighProtein$Source, sd)
```

We check the ratio between the maximum standard deviation and the minimum standard deviation:

```r
15.136416719657 / 10.9163485958752 
```

We create a Normal Q-Q plot of the residual values:

```r
qqnorm(residuals(test))
```
